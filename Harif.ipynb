#### ÿ™ÿ≠ŸàŸÑ ÿßŸÑŸÜÿµ ŸÑÿ≠ÿ±ŸàŸÅ ÿµÿ∫Ÿäÿ±Ÿáÿåÿ™ÿ≠ÿ∞ŸÅ ÿßŸÑÿßÿ≥ÿ™ŸÅŸáÿßŸÖÿå ÿ™ŸÇÿ≥ŸÖ ÿßŸÑŸÉŸÑÿßŸÖ ÿ®ÿßŸÑÿßÿ∂ÿßŸÅŸá ÿ™ÿ≠ÿ∞ŸÅ ÿßŸÑŸÉŸÑŸÖÿßÿ™ ÿßŸÑŸÑŸä ÿ≠ÿØÿØÿ™Ÿáÿß ŸÖÿ´ŸÑ what Ÿàÿ∞ŸàŸÑŸä ÿ®ÿ≠Ÿäÿ´ ŸÖÿßÿ™ÿØÿÆŸÑ ÿ∂ŸÖŸÜ ÿßŸÑÿ®ÿ≠ÿ´ . 
#### ŸäÿÆÿ™ÿßÿ± ÿßŸÑŸÉŸÑŸÖŸá ÿßŸÑŸÑŸä ŸÑŸáÿß ÿπÿØÿØ ÿßÿ≠ÿ±ŸÅ ÿßŸÉÿ®ÿ± ŸÖŸÜ Ÿ¢ 

```python
def extract_keywords(user_input: str):
    words = user_input.lower().replace("?", "").split()
    stop_words = {"is", "the", "what", "who", "when", "where", "how", "and", "or", "of", "a", "an", "in", "to", "for"}
    keywords = [w for w in words if w not in stop_words and len(w) > 2]
    return keywords
```

#### ŸÑŸÑÿ®ÿ≠ÿ´ ÿØÿßÿÆŸÑ ÿ≥ŸäŸÉÿ¥ŸÜ ÿßŸÑÿßÿ≥ÿ¶ŸÑŸá ÿßŸÑŸÑŸä ÿπÿ±ŸÅŸÜÿßŸáÿß ŸÅŸàŸÇ

```python
def search_in_qa(keywords, qa_data):
    # ÿ®ÿ≠ÿ´ ÿØŸÇŸäŸÇ: ÿßŸÑŸÖŸÅÿßÿ™Ÿäÿ≠ ÿßŸÑÿ™Ÿä ÿ™ÿ≠ÿ™ŸàŸä ŸÉŸÑ ÿßŸÑŸÉŸÑŸÖÿßÿ™ ÿßŸÑŸÖŸÅÿ™ÿßÿ≠Ÿäÿ©
    for key, answer in qa_data.items():
        key_lower = key.lower()
        if all(kw in key_lower for kw in keywords):
            return answer

    # ÿ®ÿ≠ÿ´ ÿ£ŸÇŸÑ ÿØŸÇÿ©: ÿßŸÑŸÖŸÅÿßÿ™Ÿäÿ≠ ÿßŸÑÿ™Ÿä ÿ™ÿ≠ÿ™ŸàŸä ÿ£Ÿä ŸÉŸÑŸÖÿ© ŸÖŸÅÿ™ÿßÿ≠Ÿäÿ©ÿå ŸÑŸÉŸÜ ÿ™ÿ≠ŸÇŸÇ ÿ£ŸÇÿµŸâ ÿ™ÿ∑ÿßÿ®ŸÇ
    best_match = None
    best_match_count = 0
    for key, answer in qa_data.items():
        key_lower = key.lower()
        count = sum(1 for kw in keywords if kw in key_lower)
        if count > best_match_count:
            best_match = answer
            best_match_count = count

    if best_match_count > 0:
        return best_match

    return None
```

#### ŸÑŸÑÿ®ÿ≠ÿ´ ÿØÿßÿÆŸÑ ÿ≥ŸäŸÉÿ¥ŸÜ ÿßŸÑŸÖÿπŸÑŸàŸÖÿßÿ™ ÿßŸÑÿ´ÿßÿ®ÿ™Ÿá ÿßŸÑŸÑŸä ŸÅŸäŸáÿß ÿßÿ±ŸÇÿßŸÖ ÿßŸà ÿßÿ∫ŸÑÿ®Ÿáÿß ÿßÿ±ŸÇÿßŸÖ 

```python
def search_in_stats(keywords, stats_data):
    for key, val in stats_data.items():
        if isinstance(val, dict):
            for subkey, subval in val.items():
                if any(kw in subkey.lower() for kw in keywords):
                    return f"{subkey.title()}: {subval}"
        else:
            if any(kw in key.lower() for kw in keywords):
                return f"{key.replace('_', ' ').title()}: {val}"
    return None
```

#### Ÿäÿ®ÿ≠ÿ´ ÿ®ÿßŸÑÿØÿßÿ™ÿß ÿßŸÑŸÑŸä ÿπÿ∑ŸäŸÜÿßŸá 

```python
def search_in_dataset(keywords, df):
    df_combined = df.astype(str).apply(lambda row: ' '.join(row.values).lower(), axis=1)
    for i, row in enumerate(df_combined):
        if all(kw in row for kw in keywords):
            return df.iloc[i].to_dict()
    return None
```

#### Ÿáÿ∞Ÿä ÿßŸÑŸÅŸÜŸÉÿ¥ŸÜ ŸÖÿ≥ÿ§ŸàŸÑŸá ÿßŸÜŸá Ÿäÿ™ÿßŸÉÿØ ÿßÿ∞ÿß ÿ™ŸÉŸÑŸÖ ÿ®ÿßŸÑÿßŸÜÿ¨ŸÑŸäÿ≤Ÿä ÿßŸà ÿ∫Ÿäÿ±Ÿá

```python
def is_english(text):
    # Ÿäÿ™ÿßŸÉÿØ ÿßÿ∞ÿß ŸÅŸäŸá ÿ≠ÿ±ŸàŸÅ ÿ∫Ÿäÿ± ascii 
    try:
        text.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        # ŸÑŸÑŸÅÿ≠ÿµ ÿπŸÜ ŸÉŸÑŸÖÿßÿ™ ÿßŸÜÿ¨ŸÑŸäÿ≤Ÿäÿ©
        words = re.findall(r'\b\w+\b', text.lower())
        english_words = {"the", "and", "is", "in", "you", "what", "who", "when", "where", "how"}
        count_english = sum(1 for w in words if w in english_words)
        return count_english > 0
```

#### Ÿáÿ∞Ÿä ÿßŸÑŸÅŸÜŸÉÿ¥ŸÜ ŸÑŸà ÿßŸÑŸÖÿ≥ÿ™ÿÆÿØŸÖ ÿ≥ÿ£ŸÑ ÿÆÿßÿ±ÿ¨ ŸÖŸàÿ∂ŸàÿπŸÜÿß

```python
def is_external_topic(keywords):
    external_keywords = {"basketball", "tennis", "politics", "music", "movie", "weather", "news", "technology", "stock", "economy"}
    return any(kw in external_keywords for kw in keywords)
```

```python
def random_choice(choices):
    return random.choice(choices) if choices else None
```

```python
def get_random_response(keywords):
    for kw in keywords:
        if kw in RESPONSES:
            return random_choice(RESPONSES[kw])
    return None
```

```python
def answer_QA(user_input, qa_data, stats_data, df):
    keywords = extract_keywords(user_input)
    if not keywords:
        return ("Please enter more specific keywords.", get_follow_up(['general']))

    # ŸÜÿ™ÿ≠ŸÇŸÇ ŸÖŸÜ ŸÑÿ∫ÿ™Ÿá
    if not is_english(user_input):
        return ("Sorry, I only understand English and can respond only in English.", None)

    # ŸÜÿ™ÿ≠ŸÇŸÇ ŸÖŸÜ ÿßŸÑŸÖŸàÿ∂Ÿàÿπ 
    if is_external_topic(keywords):
        # ŸÜÿ≥ŸÑŸÉ ŸÑŸá ŸàŸÜÿ±ÿ¨ÿπŸá ŸÑŸÖŸàÿ∂ÿπŸÜÿß
        return ("That's an interesting topic! But I'm really an expert on the FIFA World Cup 2022. What would you like to know about the tournament?", None)

    # ÿßŸÑÿßŸÜ ŸàŸÇÿ™ ÿßŸÑÿ®ÿ≠ÿ´ ŸÅŸä ÿßŸÑÿ´ŸÑÿßÿ´ ÿßÿ¥Ÿäÿßÿ° ÿßŸÑŸÑŸä ÿπŸÜÿØŸÜÿß
    answer = search_in_qa(keywords, qa_data)
    if not answer:
        answer = search_in_stats(keywords, stats_data)
    if not answer:
        answer = get_random_response(keywords)  # ÿ±ÿØŸàÿØ ÿ¨ÿßŸáÿ≤ÿ© ÿπÿ¥Ÿàÿßÿ¶Ÿäÿ© ŸÑŸÑŸÖŸàÿßÿ∂Ÿäÿπ ÿßŸÑŸÖÿπÿ±ŸàŸÅÿ©
    if not answer:
        answer = search_in_dataset(keywords, df)
        if answer:
            # ŸÖÿ≥ÿ§ŸàŸÑŸá ÿπŸÜ ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸâ ŸÜÿµ ŸÖÿπÿ±ŸàŸÅ ŸàŸÖŸÅŸáŸàŸÖ
            answer = "\n".join(f"{key}: {val}" for key, val in answer.items())

    # ÿßÿ∞ÿß ÿ™ŸàŸáŸÇŸÜÿß ŸàŸÖÿßÿπŸÜÿØŸÜÿß ÿßÿ¨ÿßÿ®Ÿá ŸÜÿ±ÿØ ÿ®ÿ±ÿØ ŸàŸÜŸàÿ¨Ÿá ŸÑÿ≥ÿ§ÿßŸÑ ÿßÿπŸÖŸÇ
    if not answer:
        # ŸÜÿ£ÿÆÿ∞ ŸÉŸÑŸÖÿ© ŸÖŸÅÿ™ÿßÿ≠Ÿäÿ© ÿπÿ¥Ÿàÿßÿ¶Ÿäÿ© ŸÖŸÜ ŸÉŸÑŸÖÿßÿ™ ÿßŸÑŸÖÿ≥ÿ™ÿÆÿØŸÖ ŸÑŸÑÿ≥ÿ§ÿßŸÑ ÿπŸÜŸá 
        related_kw = random_choice(keywords)
        answer = f"Oh, by the way, before I forget, I want to ask you about '{related_kw}'."

    # ÿπŸÑÿ¥ÿßŸÜ ÿ™ÿÆÿ™ÿßÿ± ÿ≥ÿ§ÿßŸÑ ÿ®ÿπÿØ ÿßŸÑÿ¨Ÿàÿßÿ®
    follow_up = get_follow_up(keywords)

    return answer, follow_up
```
